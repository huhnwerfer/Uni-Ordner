```python
import pandas as pd
import csv
import os


def filter(file):
    
    df = pd.read_csv(
			file,
			sep= ',',
			on_bad_lines='skip',
			quoting=csv.QUOTE_NONE
	)
    df = df.applymap(lambda x: x[:-1] if isinstance(x, str) and x.endswith('.') else x)
    def fix_leading_zero(x):
        if isinstance(x, str):
            if x.startswith('0') and not x.startswith('0.'):
                return x.lstrip('0') or '0'  # Falls alles Nullen, mindestens eine '0' behalten
        return x

    # Auf alle Zellen anwenden
    df = df.applymap(fix_leading_zero)
    '''duplicates = df[df.duplicated()]
    print(duplicates)''' #keine duplicate gefunden

    df.drop_duplicates()
    df.dropna()
    df.to_csv('wein_cleaned.csv', index=False)
    return df
filter("wein.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import itertools

# Lade die bereinigten Daten
df = pd.read_csv("wein_cleaned.csv")

# Sicherstellen, dass alle Spalten numerisch sind
df = df.apply(pd.to_numeric, errors='coerce')

# Entferne Zeilen mit fehlenden Werten (falls noch vorhanden)
df.dropna(inplace=True)

# Erstelle alle mÃ¶glichen Paare von numerischen Attributen
columns = df.columns
attribute_pairs = list(itertools.combinations(columns, 2))

# Plot-Stil
sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))

# Regressionsanalyse und Visualisierung
for i, (x_col, y_col) in enumerate(attribute_pairs):
    X = df[[x_col]].values
    y = df[y_col].values

    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)

    r2 = r2_score(y, y_pred)
    coef = model.coef_[0]
    intercept = model.intercept_

    # Plot
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df[x_col], y=df[y_col], color="blue", alpha=0.6)
    plt.plot(df[x_col], y_pred, color="red", linewidth=2)
    plt.title(f"{y_col} vs {x_col}\nRÂ² = {r2:.3f}, Steigung = {coef:.3f}, Intercept = {intercept:.3f}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.tight_layout()
    plt.show()

    # Textausgabe der Metriken
    print(f"Regression: {y_col} = {coef:.3f} * {x_col} + {intercept:.3f}")
    print(f"BestimmtheitsmaÃŸ (RÂ²): {r2:.3f}\n")
```
